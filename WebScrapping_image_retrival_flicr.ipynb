```python
import requests
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import time 
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Firefox()
driver.get("https://www.flickr.com/search/?text=mahindra+thar")
driver.maximize_window()
time.sleep(10)
```

```python
soup = BeautifulSoup(driver.page_source , 'html.parser')
all_div_tags = soup.find_all('div','view photo-list-photo-view awake')
print(len(all_div_tags))
```

```python
images_src = []
for everydivTag in all_div_tags:
    imgTag = everydivTag.find('img')
    imgTag = 'https:'+imgTag['src']
    images_src.append(imgTag)
```

```python
for everyimage in images_src:
    print(everyimage)
    
```

```python
import urllib.request

pic_num = 1
max_images = 30  # Maximum number of images to download

for i in images_src:
    try:
        if pic_num > max_images:  # Stop the loop after downloading 30 images
            break
        print(i)
        urllib.request.urlretrieve(i, "D:\\New folder (2)\\thar" + str(pic_num) + ".jpg")
        pic_num += 1

    except Exception as e:
        print(str(e))
```
